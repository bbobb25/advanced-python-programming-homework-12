```python
import sqlite3

connection = sqlite3.connect('books.db')

import pandas as pd

pd.options.display.max_columns = 10

cursor = connection.cursor()
```

```python
pd.read_sql('SELECT last FROM authors', connection)
```

```python
pd.read_sql('SELECT title FROM titles ORDER BY title ASC', connection)
```

```python
pd.read_sql("""SELECT first, last, title, copyright, isbn FROM authors 
INNER JOIN titles WHERE first LIKE 'Paul' ORDER BY title""", connection)
```

```python
pd.read_sql('SELECT * FROM titles', connection)
```

```python
pd.read_sql('SELECT id, first, last FROM authors', connection, index_col=['id'])
cursor = cursor.execute("""INSERT INTO authors (first, last) VALUES ('Sue', 'Red')""")

cursor.rowcount
```
